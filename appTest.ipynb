```python
from chispitapp import ChispitApp
from codegenerator import CodeGenerator
from geneticcode import genetic_code
from collections import Counter
import matplotlib.pyplot as plt
```

```python
chispis = ChispitApp()
chispis.process_files('./data/', genetic_code)
```

```python
code_generator:CodeGenerator = CodeGenerator()
generated_codes:list[list[str]] = code_generator.calculate_RSCU_scores_per_genome(chispis.get_RSCU_per_aa())
```

```python
plt.hist(dict(Counter(generated_codes[0])).values(), bins=50)
plt.yscale('log')
```
